# Purpose
This notebook serves to perform time comaprisons of each of the different modes for segmentation<br>
Ensemble <br>
MaskRCNN <br>
SAM (CPU) <br>
SAM (GPU) <br>




```python
from facet_ml.segmentation.segmenter import ImageSegmenter

from pathlib import Path
import cv2
import time
```

```python
def time_trial(is_kwargs:dict,image_list:list) -> float:
    '''
    Given kwargs for an ImageSegmenter, try running it on an image list to approximate how long it takes
    '''

    n_images = len(image_list)

    IS = ImageSegmenter(**is_kwargs)
    
    # Begin key loop
    start = time.time()
    for img_path in image_list:

        IS.input_path = str(img_path)
        IS.df
    
    stop = time.time()

    return (stop-start)/n_images
```

```python
# Get images from training
image_folder = Path(".").absolute().parent.parent / "Images" / "Training"
image_paths = list( image_folder.rglob("*") )

# Set each trial
ensemble_dict = {
    "threshold_mode" : "ensemble",
    "edge_modification" : "localthresh"
}
maskrcnn_dict = {
    "threshold_mode" : "detectron2",
    "edge_modification" : "localthresh"
}
sam_cpu_dict = {
    "threshold_mode" : "segment_anything",
    "sam_kwargs" : {
        "points_per_side": 64,
        "device" : "cpu"
    }
}
sam_gpu_dict = {
    "threshold_mode" : "segment_anything",
    "sam_kwargs" : {
        "points_per_side": 64,
        "device" : "cuda"
    }
}
```

```python
# Perform time trials
time_dict = {}

time_dict["ensemble"] = time_trial(ensemble_dict,image_list=ImageSegmenter)
time_dict["sam_cpu"] = time_trial(sam_cpu_dict,image_list=image_paths)
time_dict["sam_gpu"] = time_trial(sam_gpu_dict,image_list=image_paths)

```
