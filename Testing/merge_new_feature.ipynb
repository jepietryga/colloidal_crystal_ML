```python
import pandas as pd
import sys
sys.path.append("..")
sys.path.append("../Utility")
#from colloidal_crystal_ML 
from Utility.segmentation_utils import ImageSegmenter
import Utility.features as features
import glob
```

```python
training_image_list = glob.glob("/Users/jacobpietryga/Desktop/Academics/colloidal_crystal_ML/Images/Training/*")

df_list = []
for image in training_image_list:
    IS = ImageSegmenter(image)
    df_list.append(IS.df)

df_new = pd.concat(df_list)
```

```python
df_new = pd.concat(df_list)
df_training = pd.read_csv("/Users/jacobpietryga/Desktop/Academics/colloidal_crystal_ML/Results/training_data.csv")

print(len(df_training))

def row_adjust_filename(row:pd.Series):
    return row.Filename.split("/")[-1]

df_new_dup = df_new.copy()
df_new_dup["Filename"] = df_new_dup.apply(row_adjust_filename,axis=1)
df_training_dup = df_training.copy()
df_training_dup["Filename"] = df_training_dup.apply(row_adjust_filename,axis=1)

print(len(df_new))
check_dup = pd.concat([df_training_dup,df_new_dup])
check_dup = check_dup.drop(["Unnamed: 0.2","Unnamed: 0.1","Unnamed: 0","index","Labels",]
               ,axis=1)
check_dup = check_dup[["area","Filename","major_axis_length","minor_axis_length","perimeter"]]
check_dup = check_dup.round(2).drop_duplicates(subset=["perimeter","Filename","area"],keep=False)
print(len(check_dup))
df_training_dup[df_training_dup.Filename == "4 nM 8.bmp"].head(20)[["Region","equivalent_diameter"]]
```
