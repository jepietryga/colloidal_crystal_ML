```python
filepath = '../Data/testing_data.csv'
df = pd.read_csv(filepath)

image_name_list = np.unique(df["Filename"])
```

```python
df_fresh = pd.DataFrame()

ii = 0
for image_name in image_name_list:
    print(f'{image_name} running...')
    df_labeled = df[df["Filename"] == image_name]
    df_labeled.sort_values(by=["Region"],inplace=True)
    df_labeled = df_labeled.reset_index()
    
    nf_path = f'Results/values_{image_name[:-3]}csv'
    df_nf = pd.read_csv(nf_path)
    df_nf.sort_values(by=["Region"],inplace=True)
    df_nf = df_nf.reset_index()
    
    
    df_nf["Labels"] = df_labeled["Labels"]
    print(df_labeled["Labels"])
    print(df_nf["Labels"])
    
    df_fresh = pd.concat([df_fresh,df_nf])
```
